```python
import json
import pathlib

from loguru import logger
import pandas as pd
from tqdm.auto import tqdm
```

```python
cwd = pathlib.Path.cwd()
df_paths = cwd.parent.glob("**/*.csv")
dfs = {}

for path in df_paths:
    dfs[path.stem] = pd.read_csv(path)
    logger.info(f"Read {path.stem} with shape {dfs[path.stem].shape}")
```

```python
for key in tqdm(dfs):
    for column in tqdm(["questions", "concepts", "responses", "timestamps", "selectmasks", "is_repeat"]):
        if column in dfs[key].columns:
            dfs[key][column] = dfs[key][column].map(lambda x: [int(val) for val in x.split(",")])
            dfs[key][f"n_{column}"] = dfs[key][column].map(len)
```
